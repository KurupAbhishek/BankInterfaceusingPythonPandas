```python
import numbers
import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sb
import random as random
cust1 =["Abhishek", 11110,"01012000","1234",10000]
cust2 =["Joseph", 11100,"27091920","1111",10000]
customerdetails = pd.DataFrame(list(zip(cust1, cust2)))
customerdetails = customerdetails.T
customerdetails.columns =["Name","AccNo","DOB","PIN","AccBal"]
def create_acc():
  cusname = input("Please enter your name: ")
  custaccno = random.randint(11111,99999)
  cusaccno = custaccno
  cusdob = input("Please enter your date of birth in day month year format eg.01012000: ")
  cuspin = input("Please enter your PIN: ")
  cusbal = 0
  cust = [cusname, cusaccno, cusdob, cuspin, cusbal]
  customerdetails.loc[len(customerdetails)] = cust
  print("Your account number is", custaccno)

def show_bal():
    accno = int(input("Please enter your account number: "))
    if accno in customerdetails['AccNo'].values:
        accpin = input("Please enter your PIN: ")
        pinacc = customerdetails.loc[customerdetails['AccNo'].values == accno, 'PIN'].values[0]
        if accpin == pinacc:
            print("Your Account Balance is", customerdetails.loc[customerdetails['AccNo'].values == accno, 'AccBal'].values[0])
        else:
            print("You have entered the incorrect PIN")
    else:
        print("This account number does not exist.")

def withdraw_cash():
  accno = int(input("Please enter your account number: "))
  if accno in customerdetails['AccNo'].values:
    accpin = input("Please enter your PIN: ")
    pinacc = customerdetails.loc[customerdetails['AccNo'].values == accno, 'PIN'].values[0]
    if accpin == pinacc:
      amtwdr = int(input("Please enter the amount: "))
      amt2wdr = customerdetails.loc[customerdetails['AccNo'].values == accno, 'AccBal'].values[0]
      if amtwdr > amt2wdr:
        print("You have insufficient balance.")
      else:
        amtleft = amt2wdr - amtwdr
        print("You have withdrawn: ", amtwdr)
        customerdetails.loc[customerdetails['AccNo'].values == accno, 'AccBal'] = amtleft
    else:
      print("You have entered the incorrect PIN")
  else:
    print("This account number does not exist.")

def deposit_cash():
  accno = int(input("Please enter your account number: "))
  if accno in customerdetails['AccNo'].values:
      accpin = input("Please enter your PIN: ")
      pinacc = customerdetails.loc[customerdetails['AccNo'].values == accno, 'PIN'].values[0]
      if accpin == pinacc:
        amtdep = int(input("Please enter the amount: "))
        amt2dep = customerdetails.loc[customerdetails['AccNo'].values == accno, 'AccBal'].values[0]
        newbal = amtdep + amt2dep
        print("You have deposited: ", amtdep)
        customerdetails.loc[customerdetails['AccNo'].values == accno, 'AccBal'] = newbal
      else:
        print("You have entered the incorrect PIN")
  else:
    print("This account number does not exist.")

def atm_inter():
  atminput = int(input("Welcome please select the options. \n1. Create a bank account 2. Check account balance 3. Withdraw cash 4. Deposit money\nPress 1 to create a bank account.\nPress 2 to check account balance.\nPress 3 to withdraw money.\nPress 4 to deposit money.\n"))
  if atminput ==1:
    create_acc()
  elif atminput ==2:
    show_bal()
  elif atminput ==3:
    withdraw_cash()
  elif atminput ==4:
    deposit_cash()
  else:
    print("Please select the correct options")


atminput = []
```
